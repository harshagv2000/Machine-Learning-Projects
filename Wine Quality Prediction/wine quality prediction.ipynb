```python
def detect_outliers(df,features):
    outlier_indices = []
    
    for c in features:
        # 1st quartile
        Q1 = np.percentile(df[c],25)
        # 3st quartile
        Q3 = np.percentile(df[c],75)
        # IQR
        IQR = Q3 - Q1
        # Outlier Step
        outlier_step = IQR * 1.5
        # detect outlier and their indeces
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # store indeces 
        outlier_indices.extend(outlier_list_col)
        
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 1.5) 
    
    return multiple_outliers
```

```python
print("number of outliers detected --> ",len(data_set.loc[detect_outliers(data_set,data_set.columns[:-1])]))
data_set.loc[detect_outliers(data_set,data_set.columns[:-1])]
```

```python
data_set = data_set.drop(detect_outliers(data_set,data_set.columns[:-1]),axis = 0).reset_index(drop = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from scipy.stats import uniform ,randint
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import plot_confusion_matrix
import collections
```

```python
bins = (2, 6.5, 8)
labels = [0, 1]
data_set['quality'] = pd.cut(x = data_set['quality'], bins = bins, labels = labels)
```

```python
data_set['quality'].value_counts()
```

```python
y=data_set.quality
x=data_set.drop(['quality'],axis=1)
```

```python
test_size = 0.20
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = test_size, random_state = 206)
```

```python
scaler = StandardScaler()
X_train_sm = scaler.fit_transform(X_train) 
X_test = scaler.transform(X_test) 
```

```python
results=[]
```

```python
knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(X_train_sm, Y_train)
y_pred = knn.predict(X_test)
cm = confusion_matrix(Y_test, y_pred)

acc = accuracy_score(Y_test, y_pred)
score = knn.score(X_test, Y_test)
results.append(acc)

print("Score : ", score)
print("KNeighborsClassifier Acc : ", acc)

plot_confusion_matrix(knn, X_test, Y_test, cmap= "Greens")  
plt.show()
```

```python
print(" \t \t  KNN Classification Report")
print(classification_report(Y_test, y_pred))
```

```python
gbc = GradientBoostingClassifier(max_depth= 6, random_state=2)
gbc.fit(X_train_sm, Y_train)
y_pred_gbc = gbc.predict(X_test)
cm_aaa = confusion_matrix(Y_test, y_pred_gbc)
acc = accuracy_score(Y_test, y_pred_gbc)
score = gbc.score(X_test, Y_test)
results.append(acc)

print("Score : ", score)
print("GradientBoostingClassifier Acc : ", acc)

plot_confusion_matrix(gbc, X_test, Y_test, cmap= "binary")  
plt.show()
```

```python
svc = SVC()
svc.fit(X_train_sm, Y_train)
pred_svc = svc.predict(X_test)

cm_svc = confusion_matrix(Y_test, pred_svc)
acc = accuracy_score(Y_test, pred_svc)
score = svc.score(X_test, Y_test)
results.append(acc)

print("Score : ", score)
print("SVC Acc : ", acc)

plot_confusion_matrix(svc, X_test, Y_test, cmap= "Reds")  
plt.show()
```

```python
xgb = XGBClassifier()
xgb.fit(X_train_sm, Y_train)
pred_xgb = xgb.predict(X_test)

cm_aaa = confusion_matrix(Y_test, pred_xgb)
acc = accuracy_score(Y_test, pred_xgb)
score = xgb.score(X_test, Y_test)
results.append(acc)

print("Score : ", score)
print("XGBClassifier Acc : ", acc)

plot_confusion_matrix(xgb, X_test, Y_test, cmap= "copper")  
plt.show()
```

```python
rf = RandomForestClassifier(max_depth=18, random_state=44, bootstrap = False)
rf.fit(X_train_sm, Y_train)
y_pred_rf = rf.predict(X_test)
cm = confusion_matrix(Y_test, y_pred_rf)

acc = accuracy_score(Y_test, y_pred_rf)
score = rf.score(X_test, Y_test)
results.append(acc)

print("Score : ", score)
print("RandomForestClassifier Acc : ", acc)

plot_confusion_matrix(rf, X_test, Y_test, cmap= "pink")  
plt.show()
```

```python
df_result = pd.DataFrame({"Score":results, "ML Models":["KNN","GradientBoostingClassifier","SVC","XGBClassifier","RandomForestClassifier"]})
```
